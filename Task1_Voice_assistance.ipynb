```python


import pyttsx3
import speech_recognition as sr
import datetime
import wikipedia
import webbrowser
import os

# Use a cross-platform engine like 'espeak'
engine = pyttsx3.init("espeak")
voices = engine.getProperty("voices")
engine.setProperty("voice", voices[1].id)  # Adjust index if needed

def speak(audio):

    engine.say(audio)
    engine.runAndWait()


def wishme():
    hour = int(datetime.datetime.now().hour)
    if hour >= 0 and hour <= 12:
        speak('Good Morning!!!')
    elif hour >= 12 and hour <= 18:
        speak('Good Afternoon!!!')
    else:
        speak('Good Evening')
    speak("I Am aarohi mam. please tell me how may i help you")


def takecommand():
    r = sr.Recognizer()
    # Explicitly list available microphones and let the user choose
    mic_list = sr.Microphone.list_microphone_names()
    print("Available microphones:")
    for index, name in enumerate(mic_list):
        print(f"{index}. {name}")

    # Let the user select a microphone
    if mic_list:
         selected_index = int(input("Select microphone index: "))

         with sr.Microphone(device_index=selected_index) as source:
             print("Listening..")
             r.pause_threshold = 1
             audio = r.listen(source)
    else:
      print("no microphones found")
      return "None"
    try:
        print("Recognizing..")
        query = r.recognize_google(audio, language='en-in')
        print(f"User said:{query}\n")
    except KeyboardInterrupt:
        print("Say that again plzz..")
        return "None"
    return query


# Use __name__ to check if script is run as main program
if __name__ == "__main__":
    wishme()
    while True:
      try:
        query = takecommand().lower()
        # logic for executing tasks based on query
        if 'wikipedia' in query:
            speak('Searching Wikipedia..')
            query= query.replace("wikipedia", "")
            results = wikipedia.summary(query, sentences=2)
            speak("According to wikipedia")
            print(results)
            speak(results)

        elif 'open youtube' in query:
            webbrowser.open("youtube.com")
        elif 'open google' in query:
            webbrowser.open("google.com")



        elif 'the time' in query:
            strTime = datetime.datetime.now().strftime("%H:%M:%S")
            speak(f"Mam, the time is {strTime}")

      except KeyboardInterrupt:
            speak("Goodbye!")  # Or any other message you prefer
            break


```
